```python
import pandas as pd
import matplotlib.pyplot as plt
df= pd.read_csv("August_MySpeed_2021.csv")
# a = df.groupby(['operator']).mean()
# a
df.dropna()
# df = df.groupby(['Service_Provider']).mean()
df=df.loc[df['Download_Upload\t'] == 'download']
hh=list(df.Signal_strength)
gg=list(df.Data_Speed)
n=len(hh)
data=[]
for i in range(n) :
    data.append((hh[i],gg[i]))
    
Sig=[]
dataspee=[]
for v in data :
    if v[0] != "na" and v[1] != "na" :
        Sig.append(int(v[0]))
        dataspee.append(v[1])
del data
Sig = [int(i) for i in Sig]
amin, amax = min(Sig), max(Sig)
for i, val in enumerate(Sig):
     Sig[i] = (val-amin) / (amax-amin)
# Sig   
amin, amax = min(dataspee), max(dataspee)
for i, val in enumerate(dataspee):
     dataspee[i] = (val-amin) / (amax-amin)
# dataspee
```

```python
a = df.groupby(['Service_Provider']).mean()
```

```python
plt.figure(figsize = (10,5))
plt.bar(a['Data_Speed'].index,a['Data_Speed'].values, edgecolor = 'black');
plt.title('Average Data speed (in kbps) of \ndifferent networks', fontsize = 20)
plt.xlabel('networks', fontsize = 15)
plt.ylabel('Data speed',fontsize = 15)
plt.xticks(fontsize = 15)
```
